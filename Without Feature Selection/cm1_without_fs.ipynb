```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
cm1 = pd.read_csv(r'/content/cm1_csv(1).csv')
print('cm1 shape: ', cm1.shape)
cm1.head()
```

```python
cm1.info()
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
names = cm1.columns
d = scaler.fit_transform(cm1)
scaled_cm1 = pd.DataFrame(d, columns=names)
scaled_cm1.head()
```

```python
scaled_cm1.describe()
```

```python
scaled_cm1.corr()
```

```python
#Visualize the correlation between the features.
plt.figure(figsize=(20,12)) 
sns.heatmap(cm1.corr(),annot=True,cmap='cubehelix_r') 
plt.show()
```

```python
scaled_cm1.hist(edgecolor='black')
fig=plt.gcf()
fig.set_size_inches(30,18)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
# remove the target 
data = scaled_cm1.drop(['defects'], axis=1)
y = scaled_cm1['defects']
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=0.3, random_state=10)
```

```python
print(data.shape)
print(y.shape)
print('-------------------')
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn import metrics #for checking the model accuracy
svc = SVC()
svc.fit(x_train, y_train)
svc_pred=svc.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2) #put 2 neighbors because we have 2 labels.
knn.fit(x_train,y_train)
knn_pred = knn.predict(x_test)
```

```python
from sklearn import metrics #for checking the model accuracy
acc_svm = metrics.accuracy_score(svc_pred,y_test)
print('The accuracy of the SVM is:',acc_svm)
print('Th confusion matrix of SVM:\n ',metrics.confusion_matrix(y_test, svc_pred))
```

```python
acc_knn = metrics.accuracy_score(knn_pred,y_test)
print('The accuracy of the KNN is', acc_knn)
print('The confusion matrix of KNN:\n ',metrics.confusion_matrix(y_test, knn_pred))
```

```python
print(metrics.confusion_matrix(y_test, svc_pred))
print(metrics.classification_report(y_test, svc_pred))
```

```python
print(metrics.confusion_matrix(y_test, knn_pred))
print(metrics.classification_report(y_test, knn_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
model_score = model.score(x_test, y_test)
print(model_score)
print(metrics.confusion_matrix(y_test, y_predict))
print(metrics.classification_report(y_test, y_predict))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train) 
```

```python
test_pred = model.predict(x_test)

print(metrics.classification_report(y_test, test_pred))
print(metrics.confusion_matrix(y_test, test_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#instantiate the model
log_regression = LogisticRegression()

#fit the model using the training data
log_regression.fit(x_train,y_train)
```

```python
#define metrics
y_pred_proba = log_regression.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
